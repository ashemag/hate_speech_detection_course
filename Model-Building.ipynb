## Models for Hate Speech Detection  

For each of our 2 embedding types (tf-idf, pretrained word2vec) 
1. logistic regression (baseline)
2. convolutional neural network (CNNs)
3. long short-term memory (LSTMs) 

```python
# Load in data
import pandas as pd 
data = pd.read_pickle('data/data.pickle')
X_tfidf = pd.read_pickle('data/tfidf.pickle')
```

```python
data.shape, X_tfidf.shape
```

```python
EPOCHS = 100 
```

### Logistic Regression + TF-IDF

```python
import numpy as np 
import tensorflow as tf

# one hot encoding 
Y = tf.keras.utils.to_categorical(data['class'])
num_classes = len(Y[0])
```

```python
Y[0:10]
```

```python
def logistic_regression(flatten=False):
    model = tf.keras.models.Sequential()
    if flatten:
        model.add(tf.keras.layers.Flatten())
    model.add(tf.keras.layers.Dense(num_classes, activation=tf.nn.softmax))
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model 
```

```python
model = logistic_regression()
X = np.array(X_tfidf)
model.fit(X, Y, epochs=EPOCHS)
```

### Logistic Regression + Word2vec

```python
X = np.array(data['tweet_embedded'].to_list())

model = logistic_regression(True)
model.fit(X, Y, epochs=EPOCHS)
```

## CNN
*  A convolution is a mathematical operation that creates a set of weights, essentially creating a representation of parts of the image. This set of weights is referred to as a kernel or filter. 


```python
def cnn(filters=3, kernel_size=2, optimizer='sgd', pool_size=2, num_classes=3):
    model = tf.keras.models.Sequential()
    
    model.add(tf.keras.layers.Conv1D(filters=filters, kernel_size=kernel_size, activation='relu'))
    model.add(tf.keras.layers.MaxPool1D(pool_size))
    
    # final fully connected layer 
    model.add(tf.keras.layers.Flatten())
    model.add(tf.keras.layers.Dense(num_classes, activation=tf.nn.softmax))
    model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model 

model = cnn()
X = np.array(data['tweet_embedded'].to_list())
model.fit(X, Y, epochs=EPOCHS)
    
```

```python
model = cnn(kernel_size=1, pool_size=1)
X = np.expand_dims(np.array(X_tfidf), 1)
model.fit(X, Y, epochs=EPOCHS)
```

## LSTM

* The LSTM does have the ability to remove or add information to the cell state, carefully regulated by structures called gates.
Gates are a way to optionally let information through. They are composed out of a sigmoid neural net layer and a pointwise multiplication operation



```python
def lstm(units=4):
    model = tf.keras.models.Sequential()
    
    model.add(tf.keras.layers.LSTM(units))
    model.add(tf.keras.layers.Flatten())
    model.add(tf.keras.layers.Dense(num_classes, activation=tf.nn.softmax))
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model 
    
model = lstm()
X = np.array(data['tweet_embedded'].to_list())
model.fit(X, Y, epochs=EPOCHS)
```

```python
model = lstm()
X = np.expand_dims(np.array(X_tfidf), 1)
model.fit(X, Y, epochs=EPOCHS)
```
